Question 1
Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?








```python
df1 = ['Y2017', 'Area']
df1 = df[df1]
df1.head()
```

```python
df1.groupby('Area')['Y2017'].sum()
```

```python
df1.groupby('Area')['Y2017'].sum().idxmax()
```

Question 2:

Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
df1.groupby('Area')['Y2017'].sum().sort_values().head(7)
```

```python
df1.groupby('Area')['Y2017'].sum().sort_values().head(7).idxmax()
```

Question 3

Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?

```python
df2 = ['Element', 'Y2014', 'Y2015', 'Y2016', 'Y2017']
df2 = df[df2]
df2.head()
```

```python
df2.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017']].sum()
```

```python
df2.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017']].sum().loc['Stock Variation'].idxmax()
```
